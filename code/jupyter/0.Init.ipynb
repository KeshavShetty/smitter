# Close price info

```python
scrip_list = ["AAPL","ABB","ABBV","AEP","AGFS","AMGN","AMZN","BA","BABA","BAC","BBL","BCH","BHP","BP","BRK-A","BSAC","BUD","C","CAT","CELG","CHL","CHTR","CMCSA","CODI","CSCO","CVX",
              "D","DHR","DIS","DUK","EXC","FB","GD","GE","GOOG","HD","HON","HRG","HSBC","IEP","INTC","JNJ","JPM","KO","LMT","MA","MCD","MDT","MMM","MO","MRK","MSFT","NEE","NGG",
              "NVS","ORCL","PCG","PCLN","PEP","PFE","PG","PICO","PM","PPL","PTR","RDS-B","REX","SLB","SNP","SNY","SO","SPLP","SRE","T","TM","TOT","TSM","UL","UN","UNH","UPS","UTX",
              "V","VZ","WFC","WMT","XOM"]
# First read AAPL data as starting point and later merge with other files
all_close_price_df = pd.read_excel('D:\\projects\\datascience\\stocknet-dataset-master\\price\\kesh_processed\\AAPL.xlsx')
all_close_price_df['scrip_code'] = 'AAPL'
scrip_list.remove('AAPL')

for ascrip in scrip_list:
    file_path = 'D:\\projects\\datascience\\stocknet-dataset-master\\price\\kesh_processed\\'+ascrip+'.xlsx'    
    df = pd.read_excel(file_path)
    df['scrip_code'] = ascrip
    all_close_price_df = pd.concat([all_close_price_df, df])
```

# Merging Tweet and closing price

### Standardise or normalize date fields

```python
tweet_df['created_at'] = pd.to_datetime(tweet_df['created_at'], format = "%d/%m/%Y %H:%M")
tweet_df['parent_tweet_created_at'] = pd.to_datetime(tweet_df['parent_tweet_created_at'], format = "%d/%m/%Y %H:%M")
```

```python
tweet_df.head()
```

# Close price date

```python
all_close_price_df['Date'] = pd.to_datetime(all_close_price_df['Date'], format = "%Y-%m-%d")
```

```python
all_close_price_df['Date'].describe()
```

```python
all_close_price_df.head()
```

```python
def mapMergeDate(tweet_date, company_name) :
    if (tweet_date.hour<15) : # If hour after 3pm, then use next dat close price data
        dfQuery = '(scrip_code == \'' + company_name + '\') & (Date <=\'' + str(tweet_date) +'\')'
        #print(dfQuery)
        subset_df = all_close_price_df.loc[all_close_price_df.eval(dfQuery)]
        #print(subset_df.shape[0])
        if (subset_df.shape[0]>0):
            return max(subset_df['Date'])
        else:
            return np.nan
    else:
        dfQuery = '(scrip_code == \'' + company_name + '\') & (Date > \'' + str(tweet_date) +'\')'
        #print(dfQuery)
        subset_df = all_close_price_df.loc[all_close_price_df.eval(dfQuery)]
        #print(subset_df.shape[0])
        if (subset_df.shape[0]>0):
            return min(subset_df['Date'])
        else:
            return np.nan
    
    
#tweet_df['merge_date'] = tweet_df[['parent_tweet_created_at', 'lang']].apply(mapMergeDate) 

tweet_df['merge_date'] = tweet_df.apply(lambda x: mapMergeDate(x['created_at'], x['company_name']), axis=1)
```

```python
tweet_df.head()
```

```python
# Merge Tweet DF with  close price
merged_df = pd.merge(tweet_df, all_close_price_df,  how='inner', left_on=['company_name','merge_date'], right_on = ['scrip_code','Date'])
```
